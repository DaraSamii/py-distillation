```python
import pydistillation as distil
import numpy as np
```

# Pochon Savarit

for solving **Ponchon-Savarit** problems for designing Distillation towers we need **three** curve functions

* **yx function**
* **Liquid Enthalpy function**
* **Vapor Enthalpy function**

```python
def y(x):
    return np.sqrt(x)
```

```python
def Vapor_Enthalpy(x): 
    return -4934.444731848751 * np.power(x,3) +\
            3481.759466626407 * np.power(x,2) -\
            15273.34551506552 * x +\
            45601.84767166054
```

```python
def Liquid_Enthalpy(x):
    return 195.8253036470214 * np.power(x,3) -\
           317.772679974855 * np.power(x,3) -\
           3270.6356928062537 * x +\
           3397.860400894343
```

## Initializing Poncon-Savarit

three function must be passed to class `pydistillation.PonchonSawarit`

```python
pc = distil.PonchonSawarit(y,Vapor_Enthalpy,Liquid_Enthalpy)
```

## ploting curves
`pc.raw_plot` plots the **yx_function** ,**Vapor_Enthalpy** and **Liquid_Enthalpy** curves

```python
pc.raw_plot()
```

  ## adding operating Data
  Now we need to add following data
  * **xW**: compontent fraction in Waste,`0 <= xW <= 1`
  * **xD**: compontent fraction in Distilate,`0 <= xD <= 1`
  * **zF**: compontetn fraction in feed, `0 <= zF <= 1`
  * **F** : amount of feed in `mols`
  
  and you have to fill two of three following arguments:
  * **dHD**: delta D enthalpy 
  * **dHW**: delta W enthalpy
  * **HF**: feed enthalpy

```python
pc.addData(0.95,0.05,0.5,F=100,dHD = 100000,dHW = -50000)
```

## Plot added Data
Now we can see the feed Point, xW and xD points

```python
pc.plot_data()
```

## Compute number of Stages needed
with `pc.ComputeN` we will solve the problem with data given
and module will return data for designing distillation tower

```python
pc.ComputeN()
```

## Mulitple feed Calculations
its quite similar to simple method you just need to add second feed amount`F2`, the feed fraction `zF2` and Enthalpy of second feed`HF2`

```python
pc = distil.PonchonSawarit(y,Vapor_Enthalpy,Liquid_Enthalpy)
pc.addData(0.95,0.05,0.3,F=300,dHD = 150000,HF = 25000,zF2 = 0.6,F2 = 20, HF2=25000)
```

```python
pc.ComputeN()
```

## Open Steam

its quite similar to simple method you just need to add Open Steam amount `OS` and enthalpy of Open Steam `HOP`
if not passed Saturated steam will be considered

```python
pc = distil.PonchonSawarit(y,Vapor_Enthalpy,Liquid_Enthalpy)
pc.addData(0.95,0.05,0.45,F=100,dHD = 100000,dHW = -50000,OS=30)
pc.ComputeN()
```
