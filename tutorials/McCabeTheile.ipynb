```python
import pydistillation as distil
import numpy as np
```

# McCabe-Theile

for solving **McCabe-Theile** problems for designing Distillation towers we need **one** curve functions



```python
def y(x):
    a = 4

    return (a*x)/(1 + (a-1)*x)
```

## Initializing McCabe-Theile

three function must be passed to class `pydistillation.McCabeTheile`

```python
mc = distil.McCabeTheile(y)
```

## ploting curves
`mc.raw_plot` plots the **yx_function**  curve

```python
mc.raw_plot()
```

  ## adding operating Data
  Now we need to add following data
  * **xW**: compontent fraction in Waste,`0 <= xW <= 1`
  * **xD**: compontent fraction in Distilate,`0 <= xD <= 1`
  * **zF**: compontetn fraction in feed, `0 <= zF <= 1`
  * **F** : amount of feed in `mols`
  * **q**: quilty of input feed `0 <= q <= 1`
  * **R**: Reflux Ratio
 

```python
mc.addData(xD=0.9, xW=0.1, zF=0.5, F=100,q=0.7,R = 2)
```

## Plot added Data
Now we can see the feed Point, xW and xD points

```python
mc.plot_Data()
```

## Compute number of Stages needed
with `mc.ComputeN` we will solve the problem with data given
and module will return data for designing distillation tower

```python
mc.ComputeN()
```

## finding minimum Number of stages

```python
mc.findNmin()
```

## finiding minimum Reflux Ratio

```python
mc.findRmin()
```

## Customizing pltoing colors

```python
mc.PlotConfig(OpC='c',StageP = 'y',StageLine='m')

mc.ComputeN()
```

## Mulitple feed Calculations
its quite similar to simple method you just need to add second feed amount`F2` and the feed fraction `zF2`

```python
mc = distil.McCabeTheile(y)
mc.addData(xD=0.9, xW=0.1, zF=0.5, F=100,q=0.5,R = 2,zF2=0.6,F2=50,q2=0.6)
mc.ComputeN()
```

## Side Stream

its quite similar to simple method you just need to add Side Steam fraction `xS` and molar amount `S` and quality of the Side Stream`qs`, if not passed Saturated stream will be considered`qS = 1`

```python
mc = distil.McCabeTheile(y)
mc.addData(xD=0.9, xW=0.1, zF=0.5, F=100,q=0.5,R = 2,xS=0.7,S =25,)
mc.ComputeN()
```
